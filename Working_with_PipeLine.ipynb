# **WORKING WITH ORDINALENCODER AND LABELENCODER**
\

```python
import pandas as pd# import pandas class
import numpy as np  #import numpy class
```

```python
df=pd.read_csv('/content/pet_adoption_data.csv'); #load dataset
```

```python
df.drop(columns=['PetID'],inplace=True);
df.head() # first five rows
```

```python
df.isna().sum() # check for missing values
df.shape # shape of dataset
```

```python
df['Color'].nunique()
```

```python
x=df.iloc[:,0:11]; # independent columns
y=df.AdoptionLikelihood; #dependent column
x.head()# first five rows
```

```python
x.shape # independent features
```

```python
y.shape# dependent features
```

```python
from sklearn.model_selection import train_test_split # import sklearn train_test_split to train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) # devide train and test in 80:20
x_train.head() # first five rows
```

```python
y_test.head(10)
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder ,OrdinalEncoder # import standard scaler ,Ordinal Encoder and One Hot Encoder
from sklearn.compose import ColumnTransformer # include column transformer
ohe=OneHotEncoder(sparse=False,handle_unknown='ignore') # create a object of onehotencoder
#sparse='False' means not sparse matric formation
#handle unknown =ignore means drop unknown row in test data
ss=StandardScaler() # create a object of standard scaler
ord=OrdinalEncoder(categories=[['Small','Medium','Large']])
```

```python
len(x_test)
```

```python
x_train.head()
```

```python
# column transfer for only ordinal encoding
ct1=ColumnTransformer([
     ('tnf1',ord,[4]),
],remainder='passthrough') # rest columns are as such passed
```

```python
# column transfer for only standard. scaling
ct2=ColumnTransformer([
     ('tnf2',ss,[3,5,8,9]),
],remainder='passthrough') # rest columns are as such passed
```

```python
# column transfer for only Ordinal Enoding
ct3=ColumnTransformer([
     ('tnf3',ohe,[5,6,7]),
],remainder='passthrough') # rest columns are as such passed
```

```python
from sklearn.tree import DecisionTreeClassifier
ct4=DecisionTreeClassifier()
```

```python
from sklearn.pipeline import Pipeline
p=Pipeline([
('ct1',ct1),
('ct2',ct2),
('ct3',ct3),
('ct4',ct4),

])
p.fit(x_train,y_train)
```

```python
y_pred=p.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score # accuray score to check accuracy
```

```python
accuracy_score(y_test,y_pred) # calculate accuracy between real and predicted
```

```python
from sklearn.model_selection import cross_val_score # cross val score to check accuracy more than one time
```

```python
score=np.mean(cross_val_score(p,x_train,y_train,cv=10)) # means of 10 time accuray to get rid of bias results
score
```
